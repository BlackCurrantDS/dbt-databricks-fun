```python
# Copy dbt docs to DBFS
import sys
import shutil
from datetime import datetime

# Parameters: sys.argv[1] = source, sys.argv[2] = destination
# Example usage: source='../target'
#                destination='/dbfs/mnt/dbt_docs/dbt_sql_job/'
source_path = '../target'
destination_base = '/dbfs/mnt/dbt_docs/dbt_sql_job/'

# Create timestamped folder
run_id = datetime.now().strftime('%Y%m%d_%H%M%S')
destination_path = f"{destination_base}{run_id}/"

# Copy the entire target folder
shutil.copytree(source_path, destination_path)

print(f"Copied dbt docs from {source_path} to {destination_path}")
```
